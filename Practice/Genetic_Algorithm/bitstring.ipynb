```python
import random
p_cnt =10
max_gen = 1000

#Generate population

def Generate_Population():
    population =[]
    for i in range(p_cnt):
        population.append("".join([random.choice("01") for _ in range(5)]))
        
    return population

print(Generate_Population())
```

```python
# fitness Function

def bit_Converter(bitstr):
    val=0
    for i in range(len(bitstr)):
        k = int(bitstr[4-i])
        val += k* (2**i)
    return val

def Calculate_Fitness(bitstr):
    x = bit_Converter(bitstr)
    return 3*(x**2)+4*x+5
print(Calculate_Fitness("10100"))
```

```python
# Two point Cross over

def Crossover(p1,p2):
    child = list(p1)
    i,j = random.sample(range(len(p1)),2)
    child[i:j] = list(p2[i:j])
    return "".join(child)
    
print(Crossover("10100","11111"))
```

```python
# Wheel Selection
def roulette_wheel_selection(population, fitness_values):
    total_fitness = sum(fitness_values)
    selection_probs = [fitness / total_fitness for fitness in fitness_values]
    cumulative_probs = []
    cumulative_sum = 0.0
    for prob in selection_probs:
        cumulative_sum += prob
        cumulative_probs.append(cumulative_sum)

    def select_one():
        r = random.random()  
        for i, cp in enumerate(cumulative_probs):
            if r <= cp:
                return population[i]
            
            
    return select_one()
```

```python
# Single Point Mutation

def mutation(bitstr):
    bitstr = list(bitstr)
    i = random.randint(0,len(bitstr)-1)
    if bitstr[i] == '1':
        bitstr[i] = '0'
    else:
        bitstr[i] = '0'
    return "".join(bitstr)

print(mutation("10101"))
    
```

```python
# GA algoritm

population = Generate_Population()

for gen in range(max_gen):
    fitness = [Calculate_Fitness(i) for i in population]
    bestfit = min(fitness)
    print(f"Generation:{gen} Best Fit: {bestfit}")
    new_pop = []
    for i in range(p_cnt):
        p1 = roulette_wheel_selection(population,fitness)
        p2 = roulette_wheel_selection(population,fitness)
        
        child = Crossover(p1,p2)
        mut_rate = 0.01 +(gen/max_gen)*0.005
        if random.random()<mut_rate:
            child = mutation(child)
        new_pop.append(child)
    population = new_pop
    
fitness_values =  [Calculate_Fitness(i) for i in population]
bestfit = min(fitness)
print(f"Final Best Fit {bestfit}")
    
        
    
```
