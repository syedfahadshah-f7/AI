```python
# Environment For BFS
from collections import deque
class EnvironmentForBFS:
    
    def __init__(self,grid,row,col):
        self.grid = grid
        self.rows = row
        self.cols = col
        
    def get_percept(self):
        for i in range(self.rows):
            for j in range(self.cols):
                if self.grid[i][j] == "P":
                    return (i,j)
        return (-1,-1)
    def BFS(self,start,goal):
        came_from ={start:None}
        dir = [(0,-1),(0,1),(1,0),(-1,0)]
        q = deque([start])
        visited = set()
        visited.add(start)
        while q:
            x,y = q.popleft()
            if (x,y) == goal :
                path =[]
                current = (x,y)
                while current != None:
                    path.append(current)
                    current = came_from[current]
                path.reverse()
                return f"Goal Reached !!! Path: {path}"
            for dx,dy in dir:
                nx,ny = x+dx,y+dy
                if 0 <= nx < self.rows and 0 <= ny < self.cols and self.grid[nx][ny] != "X" and (nx,ny) not in visited:
                    came_from[(nx,ny)] = (x,y)
                    visited.add((nx,ny))
                    q.append((nx,ny))
        return "Goal Not reached"
```

```python
# Goal Based Agent for BFS

class GoalBasedAgent:
        
    def FormulateGoal(self,grid,rows,cols):
        for i in range(rows):
            for j in range(cols):
                if grid[i][j] == "T":
                    return (i,j)
        return (-1,-1)
    
    def act(self,environment,start):
        goal = self.FormulateGoal(environment.grid,environment.rows,environment.cols)
        print(f"Start: {start} Goal :{goal}")
        if goal == (-1,-1):
            return "Not valid goal\n"
        else:
            return environment.BFS(start,goal)
```

```python
# Run logic For BFS

def RunAgent(env,agent):
     percept = env.get_percept()
     if percept == (-1,-1) :
            print("Not valid percept \n")
     action = agent.act(env,percept)
     
     print(F"Action : {action}")
    

grid = [["O", "O", "X", "O" ,"T"],
["O", "X" ,"O", "O", "X"],
["P", "O", "O" ,"X", "O"],
["X", "X" ,"O" ,"O", "O"],
["O" ,"O", "O" ,"X" ,"O"]]

env = EnvironmentForBFS(grid,5,5)
agent = GoalBasedAgent()
RunAgent(env,agent)
```

```python
# Environment For DFS
from collections import deque

class EnvironmentForDFS:
    
    def __init__(self, grid, row, col):
        self.grid = grid
        self.rows = row
        self.cols = col
        self.dir = [(0, -1), (0, 1), (1, 0), (-1, 0)]
        
    def get_percept(self):
        for i in range(self.rows):
            for j in range(self.cols):
                if self.grid[i][j] == "P":
                    return (i, j)
        return (-1, -1)
    
    def RecursiveDFS(self, start, goal, visited, came_from):
        x, y = start
        if start in visited:
            return False
        visited.add(start)

        if start == goal:
            return True
        
        for dx, dy in self.dir:
            nx, ny = x + dx, y + dy
            if 0 <= nx < self.rows and 0 <= ny < self.cols and (nx, ny) not in visited:
                if self.grid[nx][ny] != "X":  
                    came_from[(nx, ny)] = (x, y)
                    if self.RecursiveDFS((nx, ny), goal, visited, came_from):
                        return True
        
        return False

    def DFS(self, start, goal):
        came_from = {start: None}
        visited = set()
        
        found = self.RecursiveDFS(start, goal, visited, came_from)
        if not found:
            return "No Path Found !!!"
        
        path = []
        current = goal
        while current is not None:
            path.append(current)
            current = came_from[current]
        
        path.reverse()
        return f"Goal Reached !!! Path: {path}"
```

```python
# Goal Based Agent for DFS

class GoalBasedAgentForDFS:
        
    def FormulateGoal(self,grid,rows,cols):
        for i in range(rows):
            for j in range(cols):
                if grid[i][j] == "T":
                    return (i,j)
        return (-1,-1)
    
    def act(self,environment,start):
        goal = self.FormulateGoal(environment.grid,environment.rows,environment.cols)
        print(f"Start: {start} Goal :{goal}")
        if goal == (-1,-1):
            return "Not valid goal\n"
        else:
            return environment.DFS(start,goal)
```

```python
# Run logic for DFS

def RunAgent(env,agent):
     percept = env.get_percept()
     if percept == (-1,-1) :
            print("Not valid percept \n")
     action = agent.act(env,percept)
     
     print(F"Action : {action}")
    

grid = [["O", "O", "X", "O" ,"T"],
["O", "X" ,"O", "O", "X"],
["P", "O", "O" ,"X", "O"],
["X", "X" ,"O" ,"O", "O"],
["O" ,"O", "O" ,"X" ,"O"]]

env = EnvironmentForDFS(grid,5,5)
agent = GoalBasedAgentForDFS()
RunAgent(env,agent)
```
